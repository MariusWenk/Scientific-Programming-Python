```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from sympy import *

x_data6 = [20.5,31.1,39.4,45.7]
y_data6 = [21.146,26.75,32.75,37.6]

s = 4
x_data61 = []
for i in range(0,s):
    x_data61.append(1/(x_data6[i]+273.15))
y_data61 = []
for i in range(0,s):
    y_data61.append(np.log(y_data6[i]))
y_dataSol61 = []
for i in range(0,s):
    y_dataSol61.append((-1/(y_data6[i]+273.15)*2357.5/8.3143))

#s = y_data6.size

#xerr = [0.1 for i in range(0,4)]
#yerr = [0.25 for i in range(0,4)]


plt.plot(x_data61,y_data61,'.',label="ln(p) nach 1/T")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data61,y_dataSol61,'.',label="ln(p) erwartet nach 1/T") 
#plt.plot(x_data6,y_dataSol6,label="theoretisch erwarteter Verlauf",linewidth=2)                        # Punkte verbunden

def fitCurve(x, A, B):
    return A * np.asarray(x) + B

pFit, pCov = curve_fit(fitCurve, x_data61, y_data61, p0=[-1,42])
plt.plot(x_data61, fitCurve(x_data61, *pFit), label="Fitkurve", color="Orange",linewidth=3)


#plt.errorbar(x_data6,y_data6,yerr,xerr,fmt='o',markersize=3,label="Dampfdruckpunkte mit Fehler",color="Black")
plt.xlabel("1/T in $1/K$")
plt.ylabel("ln(p) in $10^5 Pa$")
plt.title("Dampfdruckkurve mit ln(p) nach 1/T")
#plt.axis([0, 50, 0, 40])
plt.legend()
plt.grid(True)
plt.show()

x = Symbol('x')
A = pFit[0].round(4)
B = pFit[1].round(4)
fitCurve1 = A * x + B
fitCurve1
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from sympy import *

file5 = open("W3d_5_data.csv", encoding="charmap")
beamData5 = np.loadtxt(file5, delimiter=",")

s = beamData5[:,0].size
x_data7 = []
for i in range(0,s):
    x_data7.append(1/beamData5[i,0])
y_data7 = []
for i in range(0,s):
    y_data7.append(beamData5[i,0]*beamData5[i,1])


#xerr = [0.1 for i in range(0,s)]
#yerr = [0.25 for i in range(0,s)]


plt.plot(x_data7,y_data7,'.',label="pV Punkte")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data7,y_data7,label="pV-Kurve",linewidth=2)                        # Punkte verbunden

def fitCurve(x, A, B):
    return A * np.asarray(x) + B

pFit, pCov = curve_fit(fitCurve, x_data7, y_data7, p0=[-1,42])
plt.plot(x_data7, fitCurve(x_data7, *pFit), label="Fitkurve", color="Orange",linewidth=3)

#plt.errorbar(x_data6,y_data6,yerr,xerr,fmt='o',markersize=3,label="Dampfdruckpunkte mit Fehler",color="Black")
plt.xlabel("1/V in $1/cm³$")
plt.ylabel("pV in $10^5 Pa * cm³$")
plt.title("T > Tkrit (Messung 5) : T = 52.8 $°C$")
#plt.axis([0, 50, 0, 40])
plt.legend()
plt.grid(True)
plt.show()

x = Symbol('x')
A = pFit[0].round(4)
B = pFit[1].round(4)
fitCurve1 = A * x + B
fitCurve1
```
