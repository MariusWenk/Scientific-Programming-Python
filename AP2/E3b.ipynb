```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import ticker as ticker
from scipy.optimize import curve_fit
from sympy import *
from matplotlib import *

file = open("E3b_1_data.csv", encoding="charmap")
beamData = np.loadtxt(file, delimiter=",")

s = beamData[:,0].size
x_dataPos = beamData[:,0]
y_dataPos = beamData[:,1]
    
xerr = [0.5 for i in range(0,s)]
yerr = [0.01 for i in range(0,s)]

x_dataFit = np.arange(0,25,0.1)

#plt.plot(x_dataPos1,y_dataPos1,'.',label="Spannungsmessungen")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data7,y_data7,label="pV-Kurve",linewidth=2)                        # Punkte verbunden

def fitCurve(x, A):
    return A / np.asarray(x) 

pFit, pCov = curve_fit(fitCurve, x_dataPos, y_dataPos, p0=[-1])
plt.plot(x_dataFit, fitCurve(x_dataFit, *pFit), label="Regressionskurve durch kleinste Quadrate", color="Orange",linewidth=3)
perr = np.sqrt(np.diag(pCov))
print(perr)

plt.errorbar(x_dataPos,y_dataPos,yerr,xerr,fmt='o',markersize=3,label="Flussdichtekurve",color="Black")
plt.xlabel("Spulenlänge l in $cm$")
plt.ylabel("Flussdichte B in $mT$")
plt.title("N = 29; I = 12,5$A$")
plt.axis([9,25, 1, 4])
ax = plt.subplot(111)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(x.round(1)).replace('.',',')))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(x.round(1)).replace('.',',')))
plt.legend()
plt.grid(True)
plt.show()

x = Symbol('x')
A = pFit[0].round(4)
#B = pFit[1].round(4)
fitCurve = A / x
fitCurve
```
