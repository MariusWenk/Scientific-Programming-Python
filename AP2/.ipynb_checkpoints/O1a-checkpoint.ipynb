```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import ticker as ticker
from scipy.optimize import curve_fit
from sympy import *
from matplotlib import *

file = open("O1a_1_data.csv", encoding="charmap")
beamData = np.loadtxt(file, delimiter=",")

s = beamData[:,0].size
x_dataPos = beamData[:,0]
y_dataPos1 = beamData[:,1]
y_dataPos2 = beamData[:,2]
    
xerr = [1 for i in range(0,s)]
yerr = [0.5 for i in range(0,s)]

x_dataFit = np.arange(100,125,0.1)

#plt.plot(x_dataPos1,y_dataPos1,'.',label="Spannungsmessungen")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data7,y_data7,label="pV-Kurve",linewidth=2)                        # Punkte verbunden

def fitCurve(x, A, B):
    return A * np.asarray(x) + B

pFit1, pCov1 = curve_fit(fitCurve, x_dataPos, y_dataPos1, p0=[-1,4])
plt.plot(x_dataFit, fitCurve(x_dataFit, *pFit1), label="Regressionskurve durch kleinste Quadrate - x_1", color="Grey",linewidth=1)
pFit2, pCov2 = curve_fit(fitCurve, x_dataPos, y_dataPos2, p0=[-1,4])
plt.plot(x_dataFit, fitCurve(x_dataFit, *pFit2), label="Regressionskurve durch kleinste Quadrate - x_2", color="Blue",linewidth=1)
perr1 = np.sqrt(np.diag(pCov1))
perr2 = np.sqrt(np.diag(pCov2))
print(perr1)
print(perr2)

plt.errorbar(x_dataPos,y_dataPos1,yerr,xerr,fmt='o',markersize=3,label="Messungen mit Fehler (hier y-Fehler sehr klein) - x_1",color="Black")
plt.errorbar(x_dataPos,y_dataPos2,yerr,xerr,fmt='o',markersize=3,label="Messungen mit Fehler (hier y-Fehler sehr klein) - x_2",color="Green")
plt.xlabel("Schirmabstand a in $cm$")
plt.ylabel("Linsenpositionen x in $cm$")
plt.title("Blaues Licht")
plt.axis([100,125, 30, 90])
ax = plt.subplot(111)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(x.round(1)).replace('.',',')))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(x.round(1)).replace('.',',')))
plt.legend()
plt.grid(True)
plt.show()

print(pFit1)
print(pFit2)
x = Symbol('x')
A = pFit1[0].round(4)
B = pFit1[1].round(4)
fitCurve = A * x + B
fitCurve
```

```python
import math

file = open("O1a_1_data.csv", encoding="charmap")
beamData = np.loadtxt(file, delimiter=",")
s = beamData[:,0].size

fArr = []
delfArr = []

b = []
for i in range(0,s):
    f = round(((((beamData[i,0])**2) - ((beamData[i,2]-beamData[i,1])**2))/(4*beamData[i,0]))*10,3)
    fArr.append(f)
    delf = round(((((((beamData[i,2]-beamData[i,1])**2))/(4*((beamData[i,0])**2)))+(1/4))+(((((beamData[i,2]-beamData[i,1])))/(4*((beamData[i,0]))))*0.5))*10,3)
    delfArr.append(delf)
    b.append([beamData[i,0],f,delf])
    
plt.table(cellText=b,colLabels=["a in $cm$","f in $mm$","$\Delta$f in $mm$"],loc='top')
plt.axis("off")
plt.show()

fTrue = 0
delfTrue = 0

for i in range(0,s):
    fTrue = fArr[i] + fTrue
    delfTrue = delfArr[i] + delfTrue
    
fTrue = fTrue/s
delfTrue = delfTrue/s

print(fTrue)
print(delfTrue)
```
