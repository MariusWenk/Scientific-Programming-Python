```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from sympy import *

y_dat = np.loadtxt("W2b_Cu_isoliert_data.csv", delimiter="\t")
s = y_dat.size
y_data = []
a = 0
for i in range (0,s-a):
    y_data.append(y_dat[i])
disg = 190
dis = disg / s
disErr = 10 / s
print(disErr)
x_data = [i*dis for i in range(0,s-a)]

xerr = [disErr for i in range(1,s+1-a)]
yerr = [0.1 for i in range(1,s+1-a)]

#plot.plt(beamData[:,1],beamData[:,0],'.',label="Geschwindigkeitsmessungen")           # Einzelne Punkte ohne Fehlerbalken
#plot.plt(x_data,y_data,label="Schwingungskurve",linewidth=0.7)                        # Punkte verbunden

def fitCurve(x, A, B):
    return A * np.asarray(x) + B

pFit, pCov = curve_fit(fitCurve, x_data, y_data, p0=[-1,42])
plt.plot(x_data, fitCurve(x_data, *pFit), label="Fitkurve", color="Orange",linewidth=3)

plt.errorbar(x_data,y_data,yerr,xerr,fmt='o',markersize=3,label="Temperaturmessungen mit Fehler",color="Black")
plt.xlabel("Position x am Stab in $mm$")
plt.ylabel("Temperatur T in $°C$")
plt.title("Kupfer isoliert")
#plt.axis([0, disg, 30, 44])
plt.legend()
plt.grid(True)
plt.show()

x = Symbol('x')
A = pFit[0].round(4)
B = pFit[1].round(4)
fitCurve1 = A * x + B
fitCurve1
```
