```python
import numpy as np
from matplotlib import pyplot as plot
from scipy.optimize import curve_fit

beamData = np.loadtxt("M6a_1_data.csv", delimiter=",")

#plot.plot(beamData[:,1],beamData[:,0],'.',label="Geschwindigkeitsmessungen")
s = 627

plot.plot((beamData[:,0]),(beamData[:,1]-s),label="Schwingungskurve",linewidth=0.7)
plot.title("")
xerr = [0.1 for i in range(1,63)]
yerr = [5 for i in range(1,63)]
plot.axis([0, 3800, -100, 100])
plot.errorbar((beamData[:,0]),(beamData[:,1]-s),yerr,xerr,fmt='o',markersize=3,label="Auslenkungsmessungen",color="Black")
plot.grid(True)
plot.xlabel("Zeit t in $s$")
plot.ylabel("Bereinigte Auslenkung x in $mm$")
plot.legend()
plot.show()
```

```python
import numpy as np
from matplotlib import pyplot as plot
from scipy.optimize import curve_fit


beamData2 = np.loadtxt("M6a_2_data.csv", delimiter=",")

#plot.plot(beamData[:,1],beamData[:,0],'.',label="Geschwindigkeitsmessungen")
t = 5

z = 150
mask1 = beamData2[:, 3] == z

plot.plot((beamData2[mask1,0]),(beamData2[mask1,1]/t),label="Schneide 1",linewidth=0.7, color="Blue")
plot.plot((beamData2[mask1,0]),(beamData2[mask1,2]/t),label="Schneide 2",linewidth=0.7, color="Green")
plot.title(f"Position stationres Gewicht x = {z:.0f}$mm$")
xerr = [1 for i in range(1,18)]
yerr = [0.01/t for i in range(1,18)]
#plot.axis([0, 900, 1.7, 100])
plot.errorbar((beamData2[mask1,0]),(beamData2[mask1,1]/t),yerr,xerr,fmt='x',markersize=10,label="Schneide 1",color="Black")
plot.errorbar((beamData2[mask1,0]),(beamData2[mask1,2]/t),yerr,xerr,fmt='x',markersize=10,label="Schneide 2",color="Orange")
plot.grid(True)
plot.xlabel("Position s des Laufgewichts in $mm$")
plot.ylabel("Periodendauer T in $s$")
plot.legend()
plot.show()
```
