```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from sympy import *

file5 = open("M2a_5_data.csv", encoding="charmap")
beamData5 = np.loadtxt(file5, delimiter=",")

s = beamData5[:,0].size
lengths = [2,3.8,5.6,7.4,9.2,11]
Ts = [0,0,0,0,0,0]
for i in range(0,s):
    for j in range(0,6):
        Ts[j] += beamData5[i,j]
for j in range(0,6):
        Ts[j] /= 40
        
print(Ts)
```

```python
from sympy import *

T = Symbol('T')
D = Symbol('D')
Iz = Symbol('Iz')
mM = Symbol('mM')
a = Symbol('a')
pi = Symbol('pi')

IM1 = (((T**2)/(4*(pi**2)))*D)-Iz-(mM*(a**2))
IM2 = (((T**2)/(4*(pi**2)))*D)-Iz-(mM*(a**2))

k = 5

Ttrue = Ts[k]
Dtrue = 0.0318
Iztrue = 0.0037665
atrue = lengths[k]/100

IM1 = lambdify([T,D,Iz,mM,a,pi],IM1)
print(IM1(Ttrue,Dtrue,Iztrue,0.3408,atrue,np.pi))
```

```python
IM_diffT = IM2.diff(T)
IM_diffD = IM2.diff(D)
IM_diffIz = IM2.diff(Iz)
IM_diffmM = IM2.diff(mM)
IM_diffa = IM2.diff(a)
```

```python
k = 5

Ttrue = Ts[k]
Dtrue = 0.0318
Iztrue = 0.005728
atrue = lengths[k]/100

delD = 0.001538
delIz = 0.0002299

mistake = Abs(IM_diffT) * 0.03 +  Abs(IM_diffD) * delD + Abs(IM_diffIz) * delIz + Abs(IM_diffmM) * 0.0001 + Abs(IM_diffa) * 0.009
mistake = lambdify([T,D,Iz,mM,a,pi],mistake)
print(mistake(Ttrue,Dtrue,Iztrue,0.3408,atrue,np.pi))
```

```python
IMs = []
delIMs = [0.000649271602825342,0.0007651064916946522,0.0009125438499711468,0.001080400580863194,0.001230221028242686,0.0013895886152166022]

b = []
for i in range(0,5):
    Tss = round(Ts[i], 3)
    IM = round(IMs[i], 3)
    delIM = round(delIMs[i], 3)
    b.append([["1","2","3","4","5","6"],str(lengths[i]),str(Tss),str(IM),str(delIM)])
    
plt.table(cellText=b,colLabels=["Messung","Abstand a in $cm$","Durchschn. Periodendauer T in $s$","IM","Fehler deltaIM"],loc='top')
plt.axis("off")
plt.show()
```

```python
x_data5 = []
y_data5 = []
for i in range(0,6):
    x_data5.append((((lengths[i]/100)**2)*0.3408))
    y_data5.append((Ts[i]/(2*np.pi))**2)

xerr = [0.0001*((lengths[i]/100)**2) for i in range(0,6)]
yerr = [2*(Ts[i]/((2*np.pi)**2)*0.03) for i in range(0,6)]


#plt.plot(x_dataPos1,y_dataPos1,'.',label="Halblogarithmische Darstellung der Amplitude")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data7,y_data7,label="pV-Kurve",linewidth=2)                        # Punkte verbunden

def fitCurve(x, A, B):
    return A * np.asarray(x) + B

pFit, pCov = curve_fit(fitCurve, x_data5, y_data5, p0=[-1,4])
plt.plot(x_data5, fitCurve(x_data5, *pFit), label="Fitkurve", color="Orange",linewidth=3)

plt.errorbar(x_data5,y_data5,yerr,xerr,fmt='o',markersize=3,label="Werte mit Fehler",color="Black")
plt.xlabel("mM * a * a in $kg * m^2$")
plt.ylabel("(T/(2 * pi))^2 in $s^2$")
plt.title("")
#plt.axis([0, 50, 0, 40])
plt.legend()
plt.grid(True)
plt.show()

x = Symbol('x')
A = pFit[0].round(4)
B = pFit[1].round(4)
fitCurve1 = A * x + B
fitCurve1
```
