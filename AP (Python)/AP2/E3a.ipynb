```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from sympy import *

file = open("E3a_1_data.csv", encoding="charmap")
beamData = np.loadtxt(file, delimiter=",")


s = beamData[:,0].size
x_dataPos = beamData[:,0]
y_dataPos = beamData[:,1]

tmax = x_dataPos[s-1]
b = s-1

x_dataFit = np.arange(0,int(tmax),1)
    
xerr = [10 for i in range(0,s)]
yerr = [0.05 for i in range(0,s)]
    

#plt.plot(x_dataPos1,y_dataPos1,'.',label="Spannungsmessungen")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data7,y_data7,label="pV-Kurve",linewidth=2)                        # Punkte verbunden

def fitCurve(x, om, ph, de, xmax, A):
    return -xmax * np.exp(-de * x) * np.sin(om * x + ph) + A

pFit, pCov = curve_fit(fitCurve, x_dataPos, y_dataPos, p0=[0.005,0,0,6,0.3])
plt.plot(x_dataFit, fitCurve(x_dataFit, *pFit), label="Fitkurve durch kleinste Quadrate", color="Orange",linewidth=3)
perr = np.sqrt(np.diag(pCov))
print(perr)

plt.errorbar(x_dataPos,y_dataPos,yerr,xerr,fmt='o',markersize=3,label="Spannungsmessungen (Amplitude)",color="Black")
plt.xlabel("t in $\mu$$s$")
plt.ylabel("Spannung U in $10^-1V$")
plt.title("R = 0$\Omega$")
#plt.axis([0,14, 1.2, 1.4])
plt.legend()
plt.grid(True)
plt.show()

#x = Symbol('x')
print(pFit)
#fitCurve1 = A * x + B
#fitCurve1
```
