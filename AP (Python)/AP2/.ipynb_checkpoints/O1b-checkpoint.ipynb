```python
import math
import numpy as np
from matplotlib import pyplot as plt

file = open("O1b_1_data.csv", encoding="charmap")
beamData = np.loadtxt(file, delimiter=",")
s = beamData[:,0].size

alphArr = []
delalphArr = []
epsArr = []

b = []
for i in range(0,s):
    alph = round(beamData[i,1]-beamData[i,2],3)
    alphArr.append(alph)
    deleps = 0.2
    delalphArr.append(deleps)
    eps = round(180-alph,3)
    epsArr.append(eps)
    b.append([int(beamData[i,0]),alph,eps])
    
plt.table(cellText=b,colLabels=["Position","alpha in $°$","$\epsilon$ in $°$"],loc='top')
plt.axis("off")
plt.show()

epsTrue = 0

for i in range(0,s):
    epsTrue = epsArr[i] + epsTrue
    
epsTrue = epsTrue/s

print(epsTrue)
```

```python
import math
import numpy as np
from matplotlib import pyplot as plt

file = open("O1b_2_data.csv", encoding="charmap")
beamData = np.loadtxt(file, delimiter=",")
s = beamData[:,0].size

DelMinArr = []
nArr = []
delnArr = []
eps = np.deg2rad(59.77)
delAl = 0.2

b = []
for i in range(0,s):
    DelMinDeg = round(beamData[i,1]-beamData[i,2],3)
    DelMinArr.append(DelMinDeg)
    DelMin = np.deg2rad(DelMinDeg)
    n = round(((np.sin((DelMin+eps)/2))/(np.sin(eps/2))),3)
    nArr.append(n)
    deln = round(abs((((np.cos((DelMin+eps)/2))/(2*np.sin(eps/2)))+(((0.5*np.cos((DelMin+eps)/2)*np.sin(eps/2))-(0.5*np.sin((DelMin+eps)/2)*np.cos(eps/2)))/((np.sin(eps/2))**2)))*delAl),3)
    delnArr.append(deln)
    b.append([(beamData[i,0]),DelMinDeg,n,deln])
    
plt.table(cellText=b,colLabels=["$\lambda$ in $nm$","$\delta _{min}$ in $°$","n","$\Delta$n"],loc='top')
plt.axis("off")
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import ticker as ticker
from scipy.optimize import curve_fit
from sympy import *
from matplotlib import *

file = open("O1b_2_data.csv", encoding="charmap")
beamData = np.loadtxt(file, delimiter=",")

s = beamData[:,0].size
x_dataPos = beamData[:,0]
y_dataPos = nArr
    
#xerr = [1 for i in range(0,s)]
#yerr = [0.000001 for i in range(0,s)]

x_dataFit = np.arange(150,850,0.1)

plt.plot(x_dataPos,y_dataPos,'o',label="Messungen ohne Fehler", color="Orange")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data7,y_data7,label="pV-Kurve",linewidth=2)                        # Punkte verbunden

def fitCurve(x, n0, c ,B):
    return n0 + c / (np.asarray(x) - B)

pFit, pCov = curve_fit(fitCurve, x_dataPos, y_dataPos, p0=[3,300000000,200])
plt.plot(x_dataFit, fitCurve(x_dataFit, *pFit), label="Regressionskurve durch kleinste Quadrate", color="Black",linewidth=2)
perr = np.sqrt(np.diag(pCov))
print(perr)

#plt.errorbar(x_dataPos,y_dataPos,yerr,fmt='X',markersize=10,label="Messungen ohne Fehler",color="Blue")
plt.xlabel("Wellenlänge $\lambda$ in $nm$")
plt.ylabel("Brechungsindex n($\lambda$)")
plt.title("Quecksilberlampe")
plt.axis([300,800, 1.95, 2])
ax = plt.subplot(111)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(x.round(5)).replace('.',',')))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(x.round(1)).replace('.',',')))
plt.legend()
plt.grid(True)
plt.show()

print(pFit)
#x = Symbol('x')
#A = pFit[0].round(4)
#B = pFit[1].round(4)
#fitCurve = A * x + B
#fitCurve
```
