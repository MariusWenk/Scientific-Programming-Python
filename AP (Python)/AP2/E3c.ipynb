```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import ticker as ticker
from scipy.optimize import curve_fit
from sympy import *
from matplotlib import *

file = open("E3c_1_data.csv", encoding="charmap")
beamData = np.loadtxt(file, delimiter=",")

s = beamData[:,0].size
x_dataPos = beamData[:,1]
y_dataPos = beamData[:,0]
    
#xerr = [0.01 for i in range(0,s)]
yerr = [0.01 for i in range(0,s)]

x_dataFit = np.arange(0,1.7,0.1)

#plt.plot(x_dataPos1,y_dataPos1,'.',label="Spannungsmessungen")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data7,y_data7,label="pV-Kurve",linewidth=2)                        # Punkte verbunden

def fitCurve(x, A):
    return A * np.asarray(x)

pFit, pCov = curve_fit(fitCurve, x_dataPos, y_dataPos, p0=[-1])
plt.plot(x_dataFit, fitCurve(x_dataFit, *pFit), label="Fitkurve durch kleinste Quadrate", color="Orange",linewidth=3)
perr = np.sqrt(np.diag(pCov))
print(perr)

plt.errorbar(x_dataPos,y_dataPos,yerr,fmt='o',markersize=3,label="Stromkurve",color="Black")
plt.xlabel("Kapazität C in $\mu$$F$")
plt.ylabel("Stromstärke I in $mA$")
plt.title("Spannung U = (11,6 $\pm$ 0,02)$Hz$")
plt.axis([0,1.5, 0, 40])
ax = plt.subplot(111)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(x.round(1)).replace('.',',')))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(x.round(1)).replace('.',',')))
plt.legend()
plt.grid(True)
plt.show()

x = Symbol('x')
A = pFit[0].round(4)
#B = pFit[1].round(4)
fitCurve = A * x
fitCurve
```

```python
import math

file = open("E3c_2_data.csv", encoding="charmap")
beamData = np.loadtxt(file, delimiter=",")
s = beamData[:,0].size

AArr = []
phiArr = []
delAArr = []
delphiArr = []

b = []
for i in range(0,s):
    A = round(beamData[i,2]/beamData[i,1],3)
    AArr.append(A)
    delA = round(0.2*((1/beamData[i,1])+(beamData[i,2]/(beamData[i,1]*beamData[i,2]))),3)
    delAArr.append(delA)
    phi = round(np.rad2deg(np.arcsin(beamData[i,3]/beamData[i,4])),3)
    phiArr.append(phi)
    delphi = round(0.2*(1/np.sqrt(1-((beamData[i,3]/beamData[i,4])*(beamData[i,3]/beamData[i,4]))))*((1/beamData[i,3])+(beamData[i,3]/(beamData[i,4]*beamData[i,4]))),3)
    delphiArr.append(delphi)
    b.append([beamData[i,0],A,delA,phi,delphi])
    
plt.table(cellText=b,colLabels=["f in $Hz$","A","$\Delta$A","$\phi$ in $°$","$\Delta \phi$ in $°$ "],loc='top')
plt.axis("off")
plt.show()
```
