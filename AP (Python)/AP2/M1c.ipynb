```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from sympy import *

file1 = open("M1c_1_data.csv", encoding="charmap")
beamData1 = np.loadtxt(file1, delimiter=",")

s = beamData1[:,0].size
x_dataPos1 = beamData1[:,0]
y_dataPos1 = beamData1[:,1]
    
#xerr = [0.1 for i in range(0,s)]
yerr = [0.01 for i in range(0,s)]


#plt.plot(x_dataPos1,y_dataPos1,'.',label="Halblogarithmische Darstellung der Amplitude")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data7,y_data7,label="pV-Kurve",linewidth=2)                        # Punkte verbunden

def fitCurve(x, A, B):
    return A * np.asarray(x) + B

pFit, pCov = curve_fit(fitCurve, x_dataPos1, y_dataPos1, p0=[-1,4])
plt.plot(x_dataPos1, fitCurve(x_dataPos1, *pFit), label="Fitkurve", color="Orange",linewidth=3)

plt.errorbar(x_dataPos1,y_dataPos1,yerr,fmt='o',markersize=3,label="distanzpunkte nach periodenverschiebung",color="Black")
plt.xlabel("n")
plt.ylabel("s in $mm$")
plt.title("Ohne Lissajous")
#plt.axis([0, 50, 0, 40])
plt.legend()
plt.grid(True)
plt.show()

x = Symbol('x')
A = pFit[0].round(4)
B = pFit[1].round(4)
fitCurve1 = A * x + B
fitCurve1
```
