```python
import numpy as np
from matplotlib import pyplot as plot
from scipy.optimize import curve_fit

beamData = np.loadtxt("M4a_1_data.csv", delimiter=",")

#plot.plot(beamData[:,1],beamData[:,0],'.',label="Geschwindigkeitsmessungen")
plot.plot(beamData[:,1],beamData[:,0],label="Geschwindigkeitskurve")
plot.title("")
#plot.axis([0, 2.3, 0, 70])
plot.errorbar(beamData[:,1],beamData[:,0],beamData[:,2],fmt='o',label="Geschwindigkeitsmessungen",color="Black")
plot.grid(True)
plot.xlabel("Radius r in $mm$")
plot.ylabel("Geschwindigkeit v in $mm/s$")
plot.legend()
plot.show()
```

```python
plot.plot(beamData[:,1],beamData[:,0],label="Geschwindigkeitskurve")
plot.title("")
plot.axis([0, 2.3, 0, 70])
plot.errorbar(beamData[:,1],beamData[:,0],beamData[:,2],fmt='o',label="Geschwindigkeitsmessungen",color="Black")
plot.grid(True)
plot.xlabel("Radius r in $mm$")
plot.ylabel("Geschwindigkeit v in $mm/s$")
plot.legend()
plot.show()
```

```python
beamData2 = np.loadtxt("M4a_2_data.csv", delimiter=",")

plot.plot(beamData2[:,0],beamData2[:,1],label="Volumenstromkurve")
plot.title("")
plot.axis([0, 3.5, 0, 1.8])
plot.errorbar(beamData2[:,0],beamData2[:,1],beamData2[:,2],fmt='o',label="Volumenberechnungen aus Messungen",color="Black")
plot.grid(True)
plot.xlabel("Radius r in $mm$")
plot.ylabel("Volumenstrom (V/t) in $cm^3/s$")
plot.legend()
plot.show()
```
