```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from sympy import *

file1 = open("M1a_1_data.csv", encoding="charmap")
beamData1 = np.loadtxt(file1, delimiter=",")

s = beamData1[:,0].size
x_dataPos1 = [beamData1[i*2,0] for i in range(0,int(s/2))]
y_dataPos1 = []
for i in range(0,s):
    if beamData1[i,1] >= 0:
        y_dataPos1.append(round(np.log(beamData1[i,1]),4))
    
#xerr = [0.1 for i in range(0,s)]
#yerr = [0.25 for i in range(0,s)]


plt.plot(x_dataPos1,y_dataPos1,'.',label="Halblogarithmische Darstellung der Amplitude")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data7,y_data7,label="pV-Kurve",linewidth=2)                        # Punkte verbunden

def fitCurve(x, A, B):
    return A * np.asarray(x) + B

pFit, pCov = curve_fit(fitCurve, x_dataPos1, y_dataPos1, p0=[-1,4])
plt.plot(x_dataPos1, fitCurve(x_dataPos1, *pFit), label="Fitkurve", color="Orange",linewidth=3)

#plt.errorbar(x_data6,y_data6,yerr,xerr,fmt='o',markersize=3,label="Dampfdruckpunkte mit Fehler",color="Black")
plt.xlabel("t in $s$")
plt.ylabel("ln(A) in $Skt$")
plt.title("I = 0$A$")
#plt.axis([0, 50, 0, 40])
plt.legend()
plt.grid(True)
plt.show()

x = Symbol('x')
A = pFit[0].round(4)
B = pFit[1].round(4)
fitCurve1 = A * x + B
fitCurve1
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from sympy import *

x_data5 = [0,0.2,0.3,0.4]
y_data5 = [0.0061,0.0668,0.1557,0.2466]

s = 4
x_data51 = []
for i in range(0,s):
    x_data51.append((x_data5[i]*x_data5[i]))

#xerr = [0.1 for i in range(0,4)]
#yerr = [0.25 for i in range(0,4)]


plt.plot(x_data51,y_data5,'.',label="d nach I*I")           # Einzelne Punkte ohne Fehlerbalken
#plt.plot(x_data61,y_dataSol61,'.',label="ln(p) erwartet nach 1/T") 
#plt.plot(x_data6,y_dataSol6,label="theoretisch erwarteter Verlauf",linewidth=2)                        # Punkte verbunden

def fitCurve(x, A, B):
    return A * np.asarray(x) + B

pFit, pCov = curve_fit(fitCurve, x_data51, y_data5, p0=[-1,42])
plt.plot(x_data51, fitCurve(x_data51, *pFit), label="Fitkurve", color="Orange",linewidth=3)


#plt.errorbar(x_data6,y_data6,yerr,xerr,fmt='o',markersize=3,label="Dampfdruckpunkte mit Fehler",color="Black")
plt.xlabel("Quadrat der Dämpfungsspannung I*I in $A^2$")
plt.ylabel("Dämpfung d in $Skt/s$")
plt.title("")
#plt.axis([0, 50, 0, 40])
plt.legend()
plt.grid(True)
plt.show()

x = Symbol('x')
A = pFit[0].round(4)
B = pFit[1].round(4)
fitCurve1 = A * x + B
fitCurve1
```
